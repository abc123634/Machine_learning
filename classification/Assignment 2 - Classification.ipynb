#Load product data

```python
products = gl.SFrame('amazon_baby.gl/')
```

##Self-defined function for counting specified term

```python
def count_specific_term(dictionary, selected_term):
    if selected_term in dictionary:
        return dictionary[selected_term]
    else:
        return 0
```

##Use only selected word for sentiment analysis

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 
                  'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

# 1.Compute word count

```python
products['word_count'] = gl.text_analytics.count_words(products['review'])
```

```python
#just view some of the records to make sure we made it right.
products.head() 
```

##Count the selected terms in each review and add corresponding columns

```python
#Terms we choose to learned as features
for term in selected_words:
    print term,
```

```python
for term in selected_words:
    products[term] = products['word_count'].apply(lambda x: count_specific_term(x, term))
```

```python
#make sure we count it right
products.sort('awesome', ascending=False).head()
```

##Find the most used terms among selected words

```python
count = dict()
for term in selected_words:
    count[term] = products[term].sum()

for e in count:
    print (e, count[e])
```

# 2.Create a new sentiment analysis model using only the selected_words as features

##Define label 
We will ignore all reviews with rating = 3, since they tend to have a neutral sentiment.  Reviews with a rating of 4 or higher will be considered positive, while the ones with rating of 2 or lower will have a negative sentiment.   

```python
products = products[products['rating'] != 3] 
```

```python
products['label'] = products['rating'] >= 4
```

```python
products.sort('label')
```

##Split training set and test set

```python
training_set, test_set = products.random_split(0.8, seed=0)
```

##Train model

```python
selected_term_model = gl.logistic_classifier.create(training_set, target='label', features=selected_words)
```

```python
selected_term_model.list_fields()
```

##Inspect the coefficients to see the weighting of the terms

```python
selected_term_model['coefficients'].sort('value', ascending=False)
```

#3. Comparing the accuracy of different sentiment analysis model
compare the select_term_model with the original sentiment model which use all the terms in reviews to learn.
Also, use 'Major Class Classification' as baseline where accuracy = 27981/(27981+5326) = 0.84009 

```python
sentiment_model = gl.logistic_classifier.create(training_set, target='label', features=['word_count'])
```

```python
selected_term_model.show(view='Evaluation')
```

```python
selected_term_model.evaluate(test_set)
```

```python
sentiment_model.evaluate(test_set)
```
